```python
#Model Libraries:
import numpy as np
import pandas as pd
import os
from glob import glob
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D,BatchNormalization
from tensorflow.keras.callbacks import ReduceLROnPlateau

#Visualization Libraries:
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import itertools

# Load and prepare data
train_data = np.load('DermaMNIST_Train.npz')
test_data = np.load('DermaMNIST_Test.npz')

x_train = train_data['x_train']
y_train = train_data['y_train']
x_test = test_data['x_test']
y_test = test_data['y_test']
#x_test = x_test / 255.0  # Normalize test data
x = np.concatenate((x_train, x_test), axis=0)
y= np.concatenate((y_train, y_test), axis=0)

#Dictionary
labels = ['0',
          '1',
          '2',
          '3',
          '4',
          '5',
          '6']

lesion_type_dict = {
    '0': 'Actinic keratoses',
    '1': 'Basal cell carcinoma',
    '2': 'Benign keratosis-like lesions ',
    '3': 'Dermatofibroma',
    '4': 'Melanocytic nevi',
    '5': 'Vascular lesions',
    '6': 'Melanoma'
}

#Displaying 6 sample images from training dataset with their corresponding class labels
print("Training data shape:", x_train.shape)
print("Testing data shape:", x_test.shape)
print("Training labels shape:", y_train.shape)

# Convert one-hot encoded labels to class indices
y_indices = np.argmax(y, axis=1)
y_train_indices=np.argmax(y_train, axis=1)

#Visualize the data
fig, ax = plt.subplots(7, 5)
fig.set_figheight(15)
fig.set_figwidth(10)
for classes in range (7):
  for i, inx in enumerate(np.where(y_indices==classes)[0][:5]):
    ax[classes,i].imshow(x[inx])
    ax[classes,i].set_ylabel(labels[classes],fontsize = 20.0)
    ax[classes,i].label_outer()

#Νumber of samples of each class in the whole dataset
num_classes = []
for i in range(len(labels)):
  num_classes.append(len(np.where(y_indices==i)[0]))

dataset_samples=pd.DataFrame(num_classes,index=labels)
print("\nSamples per class (whole dataset):")
print(dataset_samples)

#Count the number of training samples for each class and display them in a DataFrame using labels as index
num_classes = []
for i in range(len(labels)):
  num_classes.append(len(np.where(y_train_indices==i)[0]))

training_samples=pd.DataFrame(num_classes,index=labels)
print("\nSamples per class (training set):")
print(training_samples)

#In this part both over-sampling and under-sampling on the training data is done to handle class imbalance
#Over-sampling: Ensures minority classes have enough representation.
#Under-sampling: Prevents the dataset from becoming too large and dominated by majority classes again.
#Between 1500-4700
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

os_dict = {0: 2000, #Increase minority classes
           1: 1500,
           2: 1500,
           3: 2500, #Increase minority classes
           4: 1500,
           5: num_classes[5],
           6: 2500} #Increase minority classes
oversample = RandomOverSampler(sampling_strategy=os_dict) # randomly duplicates existing samples from minority classes to meet the desired count
oversampled_x , oversampled_y = oversample.fit_resample(x_train.reshape(x_train.shape[0], -1), y_train_indices)
print('OS_x:%s OS_y:%s' %(oversampled_x.shape, oversampled_y.shape))

us_dict = {0: 2000,
           1: 1500,
           2: 1500,
           3: 2500,
           4: 1500,
           5: 3000, # reducing class 5
           6: 2500}

undersample = RandomUnderSampler(sampling_strategy=us_dict) #removes excess samples randomly from overrepresented classes
undersampled_x , undersampled_y = undersample.fit_resample(oversampled_x, oversampled_y)
print('US_x:%s US_y:%s' %(undersampled_x.shape, undersampled_y.shape))

#Reshape back to image format
#For the previous command image had to be flatten from shape (28, 28, 3) to (28×28×3,) for each image (a 1D vector) where 2352 = 28×28×3.
undersampled_x = undersampled_x.reshape(-1,28,28,3)

#Count samples per class in the undersampled_y
num_classes = []
for i in range(len(labels)):
  num_classes.append(len(np.where(undersampled_y==i)[0]))

undersample_y=pd.DataFrame(num_classes,index=labels)
print("\nSamples per class (undersampled_y):")
print(undersample_y)

#Transforming y_label to one-hot encoded
y_train_under = to_categorical(undersampled_y) #undersampled_y --> undersampled y_train_indices
x_train_split, x_val_split, y_train_split, y_val_split = train_test_split(
    undersampled_x, y_train_under, test_size=0.2, random_state=42, stratify=undersampled_y)

print(y_train_under.shape)
print(y_test.shape) #already one-hot encoded

#Data Augmentation (transformed versions of original images)
train_datagen = ImageDataGenerator(rescale = 1./255, #Scales pixel values from [0, 255] to [0, 1] for better neural network stability
                                  rotation_range = 10,
                                  width_shift_range = 0.2,
                                  height_shift_range = 0.2,
                                  shear_range = 0.2, #Applies shearing transformations (tilting the image)
                                  horizontal_flip = True,
                                  vertical_flip = True,
                                  fill_mode = 'wrap')

batch_size=50
train_data = train_datagen.flow(undersampled_x, y_train_under, batch_size = batch_size, seed=1)
test_datagen = ImageDataGenerator(rescale=1./255)
test_data = test_datagen.flow(x_test, y_test, batch_size = batch_size, seed=1)
val_data = test_datagen.flow(x_val_split, y_val_split, batch_size=batch_size, seed=1)

print(f'Train Set: X:%s Y:%s' %(train_data.x.shape, train_data.y.shape))
print(f'Test Set : X:%s Y:%s' %(x_test.shape, y_test.shape))

# Learning rate
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3)

#CNN model creation
num_classes = 7
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 3)),
    BatchNormalization(),
    MaxPool2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPool2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),

    Conv2D(256, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPool2D(pool_size=(2, 2)),

    Flatten(),

    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

cnn_model.summary()

# Compile model
cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model using the generator
history = cnn_model.fit(
    train_data,
    epochs=20,
    steps_per_epoch=len(train_data),
    validation_data=val_data,
    validation_steps=len(val_data),
    callbacks= reduce_lr,
)

# Evaluate the model
test_loss, test_acc = cnn_model.evaluate(test_data)
print(f"Test accuracy: {test_acc:.4f}")

# Plot CNN model
from tensorflow.keras.utils import plot_model
plot_model(cnn_model)

# Probability
y_proba = cnn_model.predict(x_test/255)
y_proba.round(2)

y_pred = np.argmax(y_proba, axis=-1)
y_pred[:10]

y_pred_name = np.array(labels)[y_pred]
y_pred_name[:10]

plt.figure(figsize=(20,10))
for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(x_test[i])
    plt.title('True label: {}, Prediction: {}'.format(labels[y_test[i].argmax()], y_pred_name[i]))
plt.tight_layout()

#Confusion matrix
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

y_test_numbers = np.array([y.argmax() for y in y_test])
cm = confusion_matrix(y_test_numbers, y_pred)
cm_display = ConfusionMatrixDisplay(cm,display_labels=labels)
fig, ax = plt.subplots(figsize=(10,10))
cm_display.plot(ax=ax)


#Classification report
print(classification_report(y_test_numbers, y_pred, target_names = labels))

```
